### MSE

image.png

```python
import numpy as np

y_1 = 21
y_2 = 10
y_3 = 14
y_hat_1 = 15
y_hat_2 = 12
y_hat_3 = 18

a = (y_1 - y_hat_1)**2
b = (y_2 - y_hat_2)**2
c = (y_3 - y_hat_3)**2

n=3
mse = (a+b+c)/n

print("MSE : {}" .format(np.round(mse,4)))
```

```python
import csv

with open('/content/abalone_mini.csv') as csvfile:
    csvreader = csv.reader(csvfile) #
    
    rows = []
    for row in csvreader: # 
        rows.append(row)

print(rows) #10   
print(len(rows))
```

#### zeros 

```python
data = np.zeros([10,11])
print(data)
```

#### enumerate

```python
for n, row in enumerate(rows):
    print(n, row)
```

####    

```python
new_rows = rows[1:]
print(new_rows)
```

####   0 1 

```python
for n, row in enumerate(new_rows):
    if row[0] == 'M' : data[n,0] = 1
    if row[0] == 'F' : data[n,1] = 1
    if row[0] == 'I' : data[n,2] = 1

    data[n, 3:] = row[1:]

print(data)
```
